The slice operator [n:m] returns the part of the string from the n’th character to the m’th character, including the first but excluding the last. In other words, start with the character at index n and go up to but do not include the character at index m. This behavior may seem counter-intuitive but if you recall the range function, it did not include its end point either.

If you omit the first index (before the colon), the slice starts at the beginning of the string. If you omit the second index, the slice goes to the end of the string.

There is no Index Out Of Range exception for a slice. A slice is forgiving and shifts any offending index to something legal.

## Exercise 6 - Mendel's First Law

```python
#From rosalind_iprb(1).txt

k = 30 #AA
m = 17 #Aa
n = 26 #aa
total = k+m+n

prob1 = (n/total)*((n-1)/(total-1)) #aa x aa
prob2 = (m/total)*(n/(total-1)) #Aa x aa; we need to account for aa x Aa as well!
prob3 = (m/total)*((m-1)/(total-1))*0.25 #Aa x Aa
                      
recessiveprob = prob1+prob2+prob3
domprob = round(1-recessiveprob, 5)

print(domprob)
```

## Exercise 7 - Computing GC Content

This one was especially difficult due to the for loop - as I had used labels as markers to chop and append the sequences before them, I could not print the last sequence. The solution seems obvious in retrospect - as the for loop only runs for the number of lines in a file, the last sequence is still stored in the seq variable, and can be retrieved with a simple append command after the for loop.

```python
seq =''
stuff = [] #Create an empty list to store values - both labels and sequences 
with open("exercises/rosalind_gc (1).txt") as file_in:
    for line in file_in:
        if line[0] == ">":
            line = line.lstrip('>')
            line = line.rstrip('\n')
            stuff.append(seq)
            stuff.append(line)
            seq = ""
        else:
            line = line.rstrip('\n')
            seq += line
    stuff.append(seq) #When it runs out of lines, it'll append the last sequence

for i in range(2, len(stuff), 2):
    GC = 0
    for nuc in stuff[i]:
        if nuc == "C" or nuc == "G":
            GC += 1
    print(stuff[i-1]) #Print label
    print(round((GC/len(stuff[i]))*100,5)) #Print sequence GC content
```

## Exercise 11 - Mortal Fibonacci Rabbits

```python
#From rosalind_fibd (1).txt

#Here, each mature pair only gives rise to a pair of offspring.

n = 87 #number of months
m = 18 #lifespan per pair - they only live for m months.
ages = [0]*m #Create a cohort table to follow the population through m months
ages[-1] = 1 #We have a pair with m months left to live

for i in range (1,n):
    new_rabbits = sum(ages[0:-1]) #We want to take a snapshot of all mature rabbits, i.e. all except newborns with m months left to live
    ages[:-1] = ages[1:] #We want to shift all rabbits left as they age - ages[0] = ages[1]...ages[-2] = ages[-1]
    ages[-1] = new_rabbits #Finally, we add the newborns to the end - we didn't do this first so we wouldn't overwrite any data

print(sum(ages)) #The total number of rabbits is simply the sum of rabbits of all ages
```

For interest, the recurrence relation is Fn = F(n-1) + F(n-2) - F(n-(m+1)); however, you need to specify the edge case when n = m, because it will go out of bounds and give funny results.

Generally what is done is that three cases are specified, after the sequence is seeded with 1,1 and months = 2. While months is less than n:
- Normal Fibonacci when months is less than m - no one has died yet
- Subtract 1 when months = m and months = m + 1 to indicate the death of the first pair
- Subsequently, when months is between m+1 and n, use the modified recurrence relation above

and then months += 1, return last item of list. 

We need the months to be strictly less than n because this is computing the next month's population during the previous month. So if n = 62, it would return the population at n = 62 when months = 61, because the first input is fed into the formula when months = 2 to give us the population for n = 3. Only then is months updated post hoc.

## Exercise 13 - Independent Alleles

```python
#from rosalind_lia (1).txt
from math import comb

k = 7 #kth generation
n = 31 #at least this number of AaBb individuals

#I did the allele table calculations and found that irrespective of genotype, if you are crossed with AaBb, you get AaBb 1/4 of the time.

offspring = pow(2,k)
prob = [pow(0.75, offspring)] #probability of 0 individuals being AaBb - 0.75^(2^k)

for i in range(1,n):
    NewCase = pow(0.75, offspring-i)*pow(0.25, i)*math.comb(offspring, i)
    prob.append(NewCase) #All cases for 1, 2...n-1 individuals being AaBb, accounting for all possible combinations
    #Consider that if n = 1, there are 2^k different offspring to be that sole AaBb - you must account for all possibilities.

hetprob = 1-sum(prob) #The idea here is that we do 1 - P(AaBb<n) = P(AaBb>=n)

print(hetprob)
```

## Exercise 17 - Consensus and Profile

```python
seq =''
seqlist = [] #Create an empty list to store values - both labels and sequences 
with open("exercises/rosalind_cons.txt") as file_in:
    for line in file_in:
        if line[0] == ">":
            line = line.lstrip('>')
            line = line.rstrip('\n')
            seqlist.append(seq)
            seqlist.append(line)
            seq = ""
        else:
            line = line.rstrip('\n')
            seq += line
    seqlist.append(seq) #When it runs out of lines, it'll append the last sequence
seqlist.pop(0) #Remove first empty space

n = len(seqlist[1])

#You cannot simply multiply [[0]*n]*4 as they will all be dependent copies of each other (Python holds references)

A = [0] * n 
C = [0] * n  
G = [0] * n 
T = [0] * n 

matrix = [A,C,G,T]

#For whatever particular nucleotide, we add 1 to the relevant position in the matrix
for i in range(1, len(seqlist), 2): 
    for j in range(0, len(seqlist[1])):
        if seqlist[i][j] == "A":
            matrix[0][j] += 1
        elif seqlist[i][j] == "C":
            matrix[1][j] += 1
        elif seqlist[i][j] == "G":
            matrix[2][j] += 1
        elif seqlist[i][j] == "T":
            matrix[3][j] += 1
            
cons = "" #Initialise consensus string and dictionary
dictionary = {0:"A", 1: "C", 2:"G", 3:"T"}

#Now, we want to vertically traverse the matrix and return the index of the maximum value, as this will inform our consensus sequence
for i in range(0, n):
    vertcompare = []
    for j in range(0,4):
        vertcompare.append(matrix[j][i]) #We append the value at the ith position of all the lists A/C/G/T in sequence
    consnuc = vertcompare.index(max(vertcompare)) #Index of maximum value will tell us what the consensus nucleotide at that position is
    cons += dictionary[consnuc]

print(cons)

print("A: ", end = "")
print(*matrix[0], sep = " ")
print("C: ", end = "")
print(*matrix[1], sep = " ")
print("G: ", end = "")
print(*matrix[2], sep = " ")
print("T: ", end = "")
print(*matrix[3], sep = " ")
```

## Exercise 20 - Finding a Shared Motif

```python
seq =''
seqlist = [] #Create an empty list to store values - both labels and sequences 
with open("exercises/rosalind_lcsm.txt") as file_in:
    for line in file_in:
        if line[0] == ">":
            line = line.lstrip('>')
            line = line.rstrip('\n')
            seqlist.append(seq)
            seqlist.append(line)
            seq = ""
        else:
            line = line.rstrip('\n')
            seq += line
    seqlist.append(seq) #When it runs out of lines, it'll append the last sequence
seqlist.pop(0) #Remove first empty space

sequences = [] #Collect all sequences
for i in range(1, len(seqlist), 2):
    sequences.append(seqlist[i])

shortest = min(sequences, key=len) 
#Let's take out the shortest sequence for analysis since the longest common substring must be from this sequence
sequences.remove(shortest)

#Generate all possible LCS candidates from the shortest string
stringcandidates = []
for i in range(0, len(shortest)):
    for j in range(len(shortest), 0, -1): 
        if j-i > 1:
            stringcandidates.append(shortest[i:j])
stringcandidates.sort(key=len, reverse=True) #Order by length so we search for the longest first

#For every candidate LCS, search every string with a window corresponding to LCS length.
for i in range(0, len(stringcandidates)):
    stringlength = len(stringcandidates[i])
    counter = 0
    for j in range(0, len(sequences)):
        stringcounter = 0
        for k in range(0, len(sequences[j])):
            if sequences[j][k:k+stringlength] == stringcandidates[i]:
                stringcounter += 1
                break #If we have a match, move on to next string.
        if stringcounter == 0:
            break #If we had no match, discard the LCS.
        else: 
            counter += stringcounter
    if counter == len(sequences): #If the number of matches in total is equal to the number of strings, we have a successful candidate
        print(stringcandidates[i])
        break #Since we search from the longest to shortest, the first successful candidate is by definition the longest of the bunch
        #We terminate the for loop here to save time and power
```
